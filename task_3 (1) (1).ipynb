1. Задача классификации.

Это такая задача, в которой имеется множество различных объектов, которые в свою очередь разделены некоторым образом на классы. Задано конечное множество объектов, про которые известно к каким классам они относятся, то есть некоторая выборка.

Про остальные объекты нет данных, то есть мы не знаем к каким классам они относятся. Нужно построить алгоритм, который бы смог, классифицировать к какому классу относится неизвестный объект.

X-это признаки, описывающие объект.
Y-это конечное число (имен, меток) классов.

Пример: Задача медицинской диагностики.
В роли объектов выступают пациенты, а признаками являются пол пациента, боль в животе и наличие неких побочных симптомов.
Порядковый признак-это тяжесть состояния (тяжелое, среднее, легкое). С помощью задачи классификации можно классифицировать заболевания.

2. Задача регрессии.

Для начала дадим определение регрессии. 
Регрессия-это зависимость между мат.ожиданием случайной величины от одной или несколько других случайных величин.
Задача регрессии заключается в том, чтобы найти такую функцию f, которая бы описывала данную зависимость.

X-характеристики объекта.
Y-это некое действительное число или числовой вектор.
Пример: Задача заключается в том, чтобы предсказать стоимость квартир (y), основываясь на характеристиках квартир (x).

3. Задача сегментации.

Сегментация изображения-попиксельное классификация изображения, то есть такое присвоение меток каждому пикселю изображения, что пиксели с одинаковыми метками имеют общие характеристики. 

Алгоритм работает с изображением как с ф-ей от двух переменных f(x,y), где x,y-это координаты пикселя.

X-матрица, описывающая изображение, где w-ширина, h-высота, n-интенсивность.

Y-множество сегментов, которые вместе покрывают всё изображение. Все пиксели в сегменте похожи по некоторому признаку (цвету, яркости, текстуре).

Пример: Спутниковые снимки. На вход подаются цветные изображения (спутниковые снимки высокого разрешения), на которых необходимо выделить области пикселей, относящихся к одному классу, например, деревья. То есть, если алгоритм найдет дерево, то он пометит его 1, а все остальное будет 0.

алгоритм построен таким образом, что он выделит пиксели, которые относятся к водоемом отдельно, и к деревьям тоже отдельно.

4. Задача детекции.

Задача заключается в том, чтобы определить к какому классу принадлежит картинка, а так же определить, где конкретно находятся искомые объекты на картинке.

Данную задачу можно разбить на две.

1.Если на картинке нужно определить местоположение одного объекта.

2.Если на картинке нужно определить местоположение нескольких объектов и их классифицировать.

1.Если на картинке одно изображение.

X-это сглаженное изображение, которое состоит из пикелей, которые в свою очередь образуют матрицу.

Y-это параметры, ограничивающего прямоугольника, где w-ширина, h-высота, а x,y-это координаты нижнего левого угла.

Пример: Предсказать ограничивающий прямоугольник одного прямоугольника. На вход поступает изображение, а с помощью параметра y мы находим объект на картинке.

2.Если на картинке несколько разных изображений.

X-сглаженное изображение, которое состоит из пикселей, которые образуют матрицу.

Y-это параметры ограничивающего прямоугольника, но так как мы хотим еще узнать какой конкретно объект изображен, то добавим значение на огр. прямоугольник к целевому вектору: 0 для первого объекта, 1 для второго объекта. <font color='red'> (непонятно, что будем делать, если наперёд неизвестно сколько объектов находится на изображении. Может объект один, а может 100) </font>

Пример: Предсказать, где находятся объекты и классифицировать их. На вход мы получаем изображение с прямоугольниками и треугольниками, а c помощью y мы размечаем наши объекты и находим местоположение, например, рамочки для треугольника и прямоугольника будут разных цветов.
