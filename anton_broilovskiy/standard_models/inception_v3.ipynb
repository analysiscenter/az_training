```python
import sys

import numpy as np
import tensorflow as tf
from pandas import ewma
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook as tqn
from IPython.display import clear_output
%matplotlib inline

sys.path.append('../..')
from dataset.models.tf import Inception_v4
from dataset.opensets import MNIST
from dataset import V,B,F
```

```python
mnist = MNIST()
```

```python
model_config = {'loss': 'ce',
                'optimizer': {'name':'Adam', 'use_locking': True},
                'inputs': dict(images={'shape': B('image_shape')},
                               labels={'classes': 10, 'transform': 'ohe', 'name': 'targets'}),
                'input_block/inputs': 'images',
                'head/units': 10,
                'output': dict(ops=['labels', 'accuracy'])}
```

```python
train_dict = {'images': B('images'),
              'labels': B('labels')}
```

```python
pipeline = (mnist.train
           .p
           .init_variable('acc', init_on_each_run=list)
           .init_model('dynamic',
                      Inception_v4,
                      'inception_v4',
                      config=model_config)
           .resize(shape=(128, 128))
           .train_model('inception_v4',
                        fetches='accuracy',
                        feed_dict=train_dict,
                        save_to=V('acc'), mode='a'))

pipeline_t = (mnist.test
           .p
           .init_variable('acc', init_on_each_run=list)
           .init_variable('prediction', init_on_each_run=list)
           .init_variable('image', init_on_each_run=list)
           .init_variable('true_value', init_on_each_run=list)
           .import_model('inception_v4', pipeline)
           .resize(shape=(128, 128))
           .predict_model('inception_v4',
                         fetches=['Inception_v4/predictions', 
                                  'Inception_v4/predicted_labels',
                                  'Inception_v4/inputs/labels',
                                  'accuracy'],
                         feed_dict=train_dict,
                         save_to=[V('image'), V('prediction'), V('true_value'), V('acc')], mode='a'))
```

```python
for i in tqn(range(400)):
    pipeline.next_batch(15, n_epochs=None, shuffle=True)
    pipeline_t.next_batch(15, n_epochs=None, shuffle=True)
    if i%10==0:
        clear_output(True)
        plt.plot(ewma(np.array(pipeline_t.get_variable('acc')[-100:]),span=10,adjust=False))
        plt.show()
```

```python
predict = (mnist.test
          .p
          .init_variable('prediction', init_on_each_run=list)
          .init_variable('image', init_on_each_run=list)
          .init_variable('true_value', init_on_each_run=list)
          .import_model('inception_v4', pipeline)
          .resize(shape=(128, 128))
          .predict_model('inception_v4',
                         fetches=['Inception_v4/inputs/images',
                                  'Inception_v4/predicted_labels',
                                  'Inception_v4/inputs/labels'],
                         feed_dict=train_dict,
                         save_to=[V('image'), V('prediction'), V('true_value')],
                         mode='a'))
```

```python
for i in tqn(range(100)):
    predict.next_batch(1, n_epochs=None, shuffle=True)
```

```python
images = predict.get_variable('image')
prediction = predict.get_variable('prediction')
true_value = predict.get_variable('true_value')

_, ax = plt.subplots(3, 4, figsize=(16, 16))
ax = ax.reshape(-1)
plot = 0
for i in np.random.randint(0, len(true_value), 12):
    ax[plot].imshow(images[i].reshape(-1,128))
    ax[plot].set_xlabel('Predict: {}, True: {}'.format(int(prediction[i][0]), int(true_value[i][0])), fontsize=18)
    plot += 1
```
