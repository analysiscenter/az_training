1.Задача классификации.

Это такая задача, в которой имеется множество различных объетов, которые в свою очередь разделены некоторым образом на классы. Задано конечное множество объектов про оторые известно, к каким классам они относятся, то есть некоторая выборка.

Про остальные объекты не известно, к каким классам они относятся. Нужно построить алгоритм, который бы смог, классфицировать  к какому классу относится неизвестный объект.


2.Задача регрессии.

Для начала дадим определение регрессии. Регрессия-это зависимость между мат.ожиданием случайной величины от одной или несколько других случайных величин.
Задача регрессии заключатеся в том, чтобы найти такую функцию f, которая бы описывала данную зависимость.

3.Задача сегментации.

Сегментация изображения-это разбиение изображения на множество покрывающих его областей.

4.Задача детекции.

Задача заключается в том, чтобы определить к какому классу принадлежит картинка, а так же определить, где конкретно находится искомый объект на картинке.
